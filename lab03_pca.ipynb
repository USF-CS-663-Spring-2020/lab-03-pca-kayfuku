```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
wdbc = pd.read_csv('../code/data/wdbc.data', header=None)
wdbc
```

```python
column_labels = ['patient_id', 'diagnosis', 'mean radius', 'mean texture', 'mean perimeter', 'mean area',
'mean smoothness', 'mean compactness', 'mean concavity',
'mean concave points', 'mean symmetry', 'mean fractal dimension',
'radius error', 'texture error', 'perimeter error', 'area error',
'smoothness error', 'compactness error', 'concavity error',
'concave points error', 'symmetry error',
'fractal dimension error', 'worst radius', 'worst texture',
'worst perimeter', 'worst area', 'worst smoothness',
'worst compactness', 'worst concavity', 'worst concave points',
'worst symmetry', 'worst fractal dimension']
wdbc.columns = column_labels
wdbc
```

```python
# Check the number of data points and number of features. 
wdbc.shape
```

```python
# Check the values of diagnosis column. 
wdbc.diagnosis.unique()
```

```python
# Check if there is a missing value. 
wdbc.diagnosis.isnull().sum() 
```

```python
wdbc['diagnosis'].replace('B', 'Benign', inplace=True)
wdbc['diagnosis'].replace('M', 'Malignant', inplace=True)
wdbc
```

```python
# Extract the label. 
wdbc_feature = wdbc.iloc[:,2:] 
wdbc_label = wdbc.iloc[:,1]
```

```python
wdbc_feature.head()
```

```python
wdbc_label.head()
```

```python
# Standardization

from sklearn.preprocessing import StandardScaler

wdbc_feature_std = StandardScaler().fit_transform(wdbc_feature)
```

```python
# PCA

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
wdbc_feature_pca = pca.fit_transform(wdbc_feature_std)
wdbc_feature_pca = pd.DataFrame(data=wdbc_feature_pca, 
                                                                columns=['PC1', 'PC2'])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure()
plt.xlabel('PC1')
plt.ylabel('PC2')
targets = ['Benign', 'Malignant']
colors = ['b', 'r']
for target, color in zip(targets, colors): 
    indicesToKeep = wdbc_label == target
    plt.scatter(wdbc_feature_pca.loc[indicesToKeep, 'PC1'], 
                        wdbc_feature_pca.loc[indicesToKeep, 'PC2'],
                        c = color
               )
```

```python
evr = pca.explained_variance_ratio_
print('\n'.join('pca {:2d}: {:0.4f} %'.format(i + 1, v*100) for i, v in enumerate(evr)))
```

```python
# Gisette dataset
```

```python
# Get the features. 
gisette_X = pd.read_csv('../code/data/gisette/gisette_train.data', sep=' ', header=None)
gisette_X
```

```python
# Drop the columns that contain a missing value. 
gisette_X = gisette_X.dropna(how='any', axis=1)
```

```python
# Get the labels. 
gisette_y = pd.read_csv('../code/data/gisette/gisette_train.labels', sep=' ', header=None)
print(type(gisette_y))
gisette_y
```

```python
# Check what it contains. 
print(type(gisette_y[0]))
gisette_y[0].unique()
```

```python
# Standardization
gisette_X_std = StandardScaler().fit_transform(gisette_X)
gisette_X_std
# gisette_X_std = gisette_X
```

```python
# PCA
pca = PCA(n_components=2)
gisette_X_pca = pca.fit_transform(gisette_X_std)
gisette_X_pca = pd.DataFrame(data=gisette_X_pca, columns=['pc1', 'pc2'])
print(type(gisette_X))
gisette_X_pca
```

```python
type(gisette_X_pca['pc1'])
```

```python
# Add the labels to the features. 
gisette_merged = pd.concat([gisette_X_pca, gisette_y], axis=1)
gisette_merged.columns = ['pc1', 'pc2', 'class']
gisette_merged
```

```python
# Replace -1 with 0 in class column. 
gisette_merged['class'].replace(-1, 0, inplace=True)
gisette_merged
```

```python
ax = None
colors = ['blue', 'green']
for i, data in gisette_merged.groupby('class'): 
#     print(type(data))
#     print(data)
    ax = data.plot.scatter(
        x='pc1', y='pc2', 
        color=colors[i], label=f'class-{i}', ax=ax
    )
```

```python
evr = pca.explained_variance_ratio_
print('\n'.join('pca {:2d}: {:0.4f} %'.format(i + 1, v*100) for i, v in enumerate(evr)))
# With standardization
# pca  1: 2.8835 %
# pca  2: 2.2590 %

# Without standardization
# pca  1: 5.1113 %
# pca  2: 3.7337 %

# Seems like 2 components are not enough. 
```

```python
# Check the explained variance ratio.  
pca = PCA().fit(gisette_X)
evr = pca.explained_variance_ratio_
print('\n'.join('pca {:2d}: {:0.4f} %'.format(i + 1, v*100) for i, v in enumerate(evr)))
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of components')
plt.ylabel('evr')
```

- It looks like the first 1000 components explain 80% variance of the data.
